```python
import simpy
import pandas as pd
import numpy as np

from src.models.facility import Facility
from src.simulation.lab_sim import ClinicalLabSimulation

stations = {
    "Accessioning": (5, 25),
    "Centrifuge": (15, 25),
    "Chemistry": (30, 30),
    "Hematology": (30, 15),
    "Cold Storage": (20, 5),
    "Waste": (5, 5),
    "Courier Dock": (0, 25)
}

facility = Facility(stations)

env = simpy.Environment()

staff = {
    "accessioning": 2,
    "techs": 3
}

machines = {
    "centrifuge": 2,
    "analyzer": 3
}

sim = ClinicalLabSimulation(
    env,
    facility,
    flows=None,
    staff=staff,
    machines=machines
)

env.process(sim.arrivals(rate=1/60))  # one per minute
env.run(until=8*3600)  # 8 hours

print("Samples processed:", len(sim.results["tat"]))
print("Avg TAT (min):", np.mean(sim.results["tat"]) / 60)
print("Avg walking distance (m):", np.mean(sim.results["distance"]))
```
